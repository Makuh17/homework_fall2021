```python
def fake_sample(max_path_length):
    rng = np.random.default_rng()
    p=1
    for i in range(15000):
        p += 2**2 
    length = rng.integers(max_path_length)
    path = rng.integers(10,size=length).tolist()
    return path

def fake_sample_q(max_path_length, q):
    path = fake_sample(max_path_length)
    q.put(path)

def fake_sample_loop(max_path_length,q):
    #print('in fake loop')
    while not e.is_set():
        #print('in loop')
        p = fake_sample(max_path_length)
        q.put(p)
```

```python
batch_size = 1000
path_length = 0
paths = []
start = time.time()
e = multiprocessing.Event()
e.clear()
with multiprocessing.Pool() as pool:
    q = multiprocessing.Manager().Queue(16)
    pool.starmap_async(fake_sample_loop, [(100,q)]*(pool._processes-1))
    while path_length < batch_size:
        path = q.get()
        paths.append(path)
        path_length += len(path)
    e.set()
end  = time.time()
print(end-start)
print(path_length, len(path))
```

```python
import gym
import multiprocessing
import time
from cs285.policies.MLP_policy import MLPPolicyPG
import cs285.infrastructure.utils as utils
import concurrent.futures
import numpy as np
import torch
```

```python
def sample_trajectory(env, policy, max_path_length, render=False, render_mode=('rgb_array')):
    # initialize env for the beginning of a new rollout
    ob = env.reset()  # HINT: should be the output of resetting the env
    # init vars
    obs, acs, rewards, next_obs, terminals, image_obs = [], [], [], [], [], []
    steps = 0
    while True:
        # render image of the simulated env
        if render:
            if 'rgb_array' in render_mode:
                if hasattr(env, 'sim'):
                    image_obs.append(env.sim.render(camera_name='track', height=500, width=500)[::-1])
                else:
                    image_obs.append(env.render(mode=render_mode))
            if 'human' in render_mode:
                env.render(mode=render_mode)
                time.sleep(env.model.opt.timestep)
        # use the most recent ob to decide what to do
        obs.append(ob)
        ac = policy.get_action(ob)  # HINT: query the policy's get_action function
        ac = ac[0]
        acs.append(ac)

        # take that action and record results
        ob, rew, done, _ = env.step(ac)

        # record result of taking that action
        steps += 1
        next_obs.append(ob)
        rewards.append(rew)

        rollout_done = 1 if done or steps >= max_path_length else 0
        terminals.append(rollout_done)

        if rollout_done:
            break
    return Path(obs, image_obs, acs, rewards, next_obs, terminals)
```

```python
#env = gym.make('Hopper-v2')
env = gym.make('LunarLanderContinuous-v2')
discrete = isinstance(env.action_space, gym.spaces.Discrete)
# Are the observations images?
img = len(env.observation_space.shape) > 2
ob_dim = env.observation_space.shape if img else env.observation_space.shape[0]
ac_dim = env.action_space.n if discrete else env.action_space.shape[0]
n_l = 2
size = 64

policy = MLPPolicyPG(ac_dim, ob_dim, n_l, size)

batch_size = 20000
ep_len = 1000

seed = 5
np.random.seed(seed)
torch.manual_seed(seed)
env.seed(seed)
```
